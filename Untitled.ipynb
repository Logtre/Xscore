```python
# -*- coding:utf-8 -*-
from sklearn import linear_model
import matplotlib.pyplot as plt

from draw_chart import DrawChart
from util import CountRecord
from file_io import FileIO

draw_chart = DrawChart()
count_rec = CountRecord()
file_io = FileIO()

# ファイルオープン処理
file = file_io.open_file_as_pandas("./data/out/20181005.sales.csv","utf-8")

clf = linear_model.LinearRegression()

# 説明変数に "施術時間" を利用
X = file.loc[:, ['施術時間']].as_matrix()

# 目的変数に "現金外支払合計" を利用
Y = file['現金外支払合計'].as_matrix()

# 予測モデルを作成
clf.fit(X, Y)

# 回帰係数
print(clf.coef_)

# 切片 (誤差)
print(clf.intercept_)

# 決定係数
print(clf.score(X, Y))

# 散布図
plt.scatter(X, Y)
# 回帰直線
plt.plot(X, clf.predict(X))
```
